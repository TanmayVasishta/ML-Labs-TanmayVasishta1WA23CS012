```
import pandas as pd
import numpy as np
df=pd.read_csv('housing.csv')
```

```
df.info()
```

```
# 3. Display statistical information of all numerical columns
print("=" * 60)
print("STATISTICAL INFORMATION - NUMERICAL COLUMNS")
print("=" * 60)
print(df.describe())
print("\n")
```

```
# 4. Display count of unique labels for 'ocean_proximity' column
print("=" * 60)
print("UNIQUE LABELS COUNT - OCEAN PROXIMITY")
print("=" * 60)
print("Value counts for 'ocean_proximity':")
print(df['ocean_proximity'].value_counts())
print(f"\nTotal unique labels: {df['ocean_proximity'].nunique()}")
print("\n")

# 5. Display attributes with missing values count greater than zero
print("=" * 60)
print("COLUMNS WITH MISSING VALUES")
print("=" * 60)
missing_values = df.isnull().sum()
columns_with_missing = missing_values[missing_values > 0]
if len(columns_with_missing) > 0:
    print("Columns with missing values:")
    print(columns_with_missing)
else:
    print("No columns have missing values.")
print("\n")
```
