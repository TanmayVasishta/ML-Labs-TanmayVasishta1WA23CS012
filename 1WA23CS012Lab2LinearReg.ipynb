```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Load dataset
df = pd.read_csv("canada_per_capita_income.csv")

# Independent and dependent variables
X = df[['year']]
y = df['per capita income (US$)']

# Create and train model
model = LinearRegression()
model.fit(X, y)

# Predict income for year 2020
pred_2020 = model.predict([[2020]])
print("Predicted per capita income in 2020:", pred_2020[0])

# Visualization
plt.scatter(X, y, color='blue')
plt.plot(X, model.predict(X), color='red')
plt.xlabel("Year")
plt.ylabel("Per Capita Income")
plt.show()
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

df = pd.read_csv("salary.csv")
print(df.isna().sum())
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Load dataset
df = pd.read_csv("1000_Companies.csv")

# One-hot encoding for State
df = pd.get_dummies(df, drop_first=True)

X = df.drop('Profit', axis=1)
y = df['Profit']

model = LinearRegression()
model.fit(X, y)

# Prediction
pred_profit = model.predict([[91694.48, 515841.3, 11931.24, 1, 0]])
print("Predicted Profit:", pred_profit[0])
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Load dataset
df = pd.read_csv("hiring.csv")

# Fix missing experience
df['experience'] = df['experience'].fillna(0)

# Convert word numbers to numeric
word_to_num = {
    'zero':0, 'one':1, 'two':2, 'three':3, 'four':4,
    'five':5, 'six':6, 'seven':7, 'eight':8,
    'nine':9, 'ten':10, 'eleven':11, 'twelve':12
}
df['experience'] = df['experience'].apply(
    lambda x: word_to_num[x] if isinstance(x, str) else x
)

# ðŸ”§ Fix missing test & interview scores using mean
df['test_score(out of 10)'] = df['test_score(out of 10)'].fillna(
    df['test_score(out of 10)'].mean()
)

df['interview_score(out of 10)'] = df['interview_score(out of 10)'].fillna(
    df['interview_score(out of 10)'].mean()
)

# Independent variables
X = df[['experience',
        'test_score(out of 10)',
        'interview_score(out of 10)']]

# Target
y = df['salary($)']

# Train model
model = LinearRegression()
model.fit(X, y)

# Predictions required by lab
print("Salary (2 yrs, 9 test, 6 interview):",
      model.predict([[2, 9, 6]])[0])

print("Salary (12 yrs, 10 test, 10 interview):",
      model.predict([[12, 10, 10]])[0])
```
